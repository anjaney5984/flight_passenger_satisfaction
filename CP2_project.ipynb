## Feature Selection Using Chi square and anova

```python
import pandas as pd
```

```python
df=pd.read_csv(r"D:\train.csv")
df.head()
```

```python
df=df.drop(columns=['Unnamed: 0','id'])
```

```python
df['Arrival Delay in Minutes'].fillna(df['Arrival Delay in Minutes'].mean(),inplace=True)
```

```python
    df_x=df.iloc[::,df.columns!='satisfaction']
    df_x.columns
    df_y=df.iloc[:,-1]
```

```python
from scipy.stats import chi2_contingency
p_values=[]
for i in df_x.columns:
    a=pd.crosstab(df[i],df_y)
    p=chi2_contingency(a)[1]
    p_values.append(p)
p_values

```

```python
df1 = pd.DataFrame()
df1['cols'] = df_x.columns
df1['values'] = p_values
df1
```
